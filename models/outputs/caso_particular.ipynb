```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('ejemplo.xlsx')
df = df.drop('Unnamed: 0',axis=1)
df.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(
    data=df.corr(),
    cmap=sns.diverging_palette(20,230,as_cmap=True),
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    annot=True,
)
```

```python
plt.figure(figsize=(15,4))
sns.scatterplot(
    data=df,
    x='Nivel Inicial [m]',
    y='Q Obj [m³/s]',
    alpha=1/30
)
```

```python
plt.figure(figsize=(15,4))
sns.scatterplot(
    data=df,
    x='Nivel Inicial [m]',
    y='Nf [m]',
    alpha=1/5
)
```
